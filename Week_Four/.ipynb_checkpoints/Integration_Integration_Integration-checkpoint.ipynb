# 7.1 - Integration-by Parts 

So, let me teach you one of the most useful results in Calculus ever.  If we think back, we should remember the product rule for differentiation

$$
\frac{d}{dx}(uv) = v(x)\frac{d}{dx}u(x) + u(x)\frac{d}{dx}v(x)
$$

Now what if I integrated all of this with respect to $x$?

$$
\int \frac{d}{dx}(uv) dx = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

Well, using the Fundamental Theorem, the left-hand side  of all of this must give us 

$$
u(x)v(x) + C = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

The way we are going to write this though is 

$$
\int u(x)\frac{d}{dx}v(x) dx = u(x)v(x) - \int v(x)\frac{d}{dx}u(x) dx
$$

leaving the constant of integration $C$ to be implied.  This formula sets up a technique called *integration-by parts*.

### Definite Integrals

Note, we can easily port the integration by parts formula over to definite integrals, where it now looks like 

$$
\int_{a}^{b} u(x)\frac{d}{dx}v(x) dx = \left. u(x)v(x)\right|_{a}^{b} - \int_{a}^{b} v(x)\frac{d}{dx}u(x) dx
$$

**Problem**:  Find (remember $d/dx \tan^{-1}(x) = 1/(1+x^{2}))$
$$
\int_{0}^{1} \tan^{-1}(x) dx = \int_{0}^{1} \tan^{-1}(x) \cdot 1 dx
$$



## 7.2 Trignometric Integrals

Key facts we need here (and yes, you need to have this memorized so you don't waste time trying to look it up)

> $\cos^{2}(x) + \sin^{2}(x)=1$

> $\frac{d}{dx}\sin(x) = \cos(x)$,  $\frac{d}{dx}\cos(x) = -\sin(x)$

Also, we are going to use "usubs" left and right.  To wit 

**Problem**: Find $\int \cos^{3}(x)dx$

So, this starts with a flat out trick:

$$
\cos^{3}(x) = \cos^{2}(x) \cos(x) = \left(1-\sin^{2}(x)\right)\cos(x)
$$

Then we see that 

\begin{align*}
\int \left(1-\sin^{2}(x)\right)\cos(x)dx = & \int \cos(x) dx - \int \sin^{2}(x)\cos(x)dx \\
\text{how I learned it growing up }= & \int \cos(x) dx - \int \sin^{2}(x)d\left(\sin(x)\right)\\
\text{or } = & \int \cos(x) dx - \int u^{2}du, ~ u = \sin(x)\\
= & \sin(x) - \frac{1}{3}\sin^{3}(x) + C
\end{align*}

**Problem**: This trickery can go in weird ways.  Find

\int \sin^{3}(x)\cos^{2}(x)dx
$$
