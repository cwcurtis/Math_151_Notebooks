# 7.1 - Integration-by Parts 

So, let me teach you one of the most useful results in Calculus ever.  If we think back, we should remember the product rule for differentiation

$$
\frac{d}{dx}(uv) = v(x)\frac{d}{dx}u(x) + u(x)\frac{d}{dx}v(x)
$$

Now what if I integrated all of this with respect to $x$?

$$
\int \frac{d}{dx}(uv) dx = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

Well, using the Fundamental Theorem, the left-hand side  of all of this must give us 

$$
u(x)v(x) + C = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

The way we are going to write this though is 

$$
\int u(x)\frac{d}{dx}v(x) dx = u(x)v(x) - \int v(x)\frac{d}{dx}u(x) dx
$$

leaving the constant of integration $C$ to be implied.  This formula sets up a technique called *integration-by parts*.

Let's see more examples.  For the next two keep in mind the central fact 

> $\frac{d}{dx}e^{x} = e^{x}$.

$$

$$

### Definite Integrals

Note, we can easily port the integration by parts formula over to definite integrals, where it now looks like 

$$
\int_{a}^{b} u(x)\frac{d}{dx}v(x) dx = \left. u(x)v(x)\right|_{a}^{b} - \int_{a}^{b} v(x)\frac{d}{dx}u(x) dx
$$

$$



## 7.2 Trignometric Integrals

Key facts we need here (and yes, you need to have this memorized so you don't waste time trying to look it up)

> $\cos^{2}(x) + \sin^{2}(x)=1$

> $\frac{d}{dx}\sin(x) = \cos(x)$,  $\frac{d}{dx}\cos(x) = -\sin(x)$

Also, we are going to use "usubs" left and right.  To wit 

**Problem**: Find $\int \cos^{3}(x)dx$

So, this starts with a flat out trick:

$$
\cos^{3}(x) = \cos^{2}(x) \cos(x) = \left(1-\sin^{2}(x)\right)\cos(x)
$$

Then we see that 

\begin{align*}
\int \left(1-\sin^{2}(x)\right)\cos(x)dx = & \int \cos(x) dx - \int \sin^{2}(x)\cos(x)dx \\
\text{how I learned it growing up }= & \int \cos(x) dx - \int \sin^{2}(x)d\left(\sin(x)\right)\\
\text{or } = & \int \cos(x) dx - \int u^{2}du, ~ u = \sin(x)\\
= & \sin(x) - \frac{1}{3}\sin^{3}(x) + C
\end{align*}

**Problem**: This trickery can go in weird ways.  Find 

$$
\int \sin^{3}(x)\cos^{2}(x)dx
$$

Hint: 

$$
\int \sin^{3}(x)\cos^{2}(x)dx = \int \sin(x)\sin^{2}(x)\cos^{2}(x)dx
$$

### Double Angle Formula tricks

> $\cos^{2}(x) = \frac{1+\cos(2x)}{2}$

> $\sin^{2}(x) = \frac{1-\cos(2x)}{2}$

So then we see for example

\begin{align*}
\int_{0}^{\pi}\sin^{2}(x) dx = & \frac{1}{2}\int_{0}^{\pi}\left(1 - \cos(2x)\right) dx\\
= & \frac{1}{2}\left(\int_{0}^{\pi}dx - \int_{0}^{\pi}\cos(2x) dx \right)\\
= & \frac{1}{2}\left.\left(x - \frac{1}{2}\sin(2x) \right)\right|_{0}^{\pi}\\
= & \frac{\pi}{2}
\end{align*}

**Problem**: Find 

\begin{align*}
\int \sin^{4}(x) dx = & \int \left(\sin^{2}(x)\right)^{2}dx\\
= & \frac{1}{4}\int \left(1 - 2\cos(2x) + \cos^{2}(2x) \right) dx\\
 = & \frac{1}{4}\int \left(1 - 2\cos(2x) + \frac{1}{2}\left(1+\cos(4x)\right) \right) dx
\end{align*}

### $\tan(x)$ and $\sec(x)$ Tricks

So here, we need to remember a few key facts

> $\frac{d}{dx}\tan(x) = \sec^{2}(x)$, $\frac{d}{dx}\text{sec}(x) = \sec(x)\tan(x)$

> $1+\tan^{2}(x) = \sec^{2}(x)$

So again, we see that $\tan(x)$ and $\sec(x)$ are tied to one another in a similar fashion as we see from $\cos(x)$ and $\sin(x)$.

For example then 

\begin{align*}
\int \tan^{6}(x) \sec^{4}(x) dx = & \int \tan^{6}(x)\sec^{2}(x)\sec^{2}(x)dx \\
= & \int \tan^{6}(x)\left(1+\tan^{2}(x) \right) d\left(\tan(x) \right), ~ d(\tan(x)) = \sec^{2}(x)dx\\
= & \frac{1}{7}\tan^{7}(x) + \frac{1}{9}\tan^{9}(x) + C
\end{align*}

**Problem**: Find

\begin{align*}
\int \tan^{5}(x)\sec^{7}(x) dx = & \int \tan^{4}(x)\sec^{6}(x)\sec(x)\tan(x)dx\\ 
= & \int \left(\sec^{2}(x)-1\right)^{2}\sec^{6}(x)d(\sec(x))
\end{align*}

# 7.3 - Trig Substitutions 

What to do with integrals of the form 

$$
\int \sqrt{1-x^{2}} dx, ~ \int \sqrt{1+x^{2}} dx, ~ \int \sqrt{x^{2}-1} dx
$$

Well, we have three fundamental identities

$$
1-\cos^{2}(t) = \sin^{2}(t), ~ 1+ \tan^{2}(t)= \sec^{2}(t), ~ \sec^{2}(t)-1=\tan^{2}(t)
$$

So in the first case, let us use $x = \cos(t)$ so that $dx = -\sin(t)dt$ and we have 

\begin{align*}
\int \sqrt{1-x^{2}} dx = & -\int \sqrt{1-\cos^{2}(t)}\sin(t) dt\\
= & -\int \sin^{2}(t) dt \\
= & -\frac{1}{2}\int \left( 1 - \cos(2t)\right) dt\\
= & -\frac{1}{2}\left(t - \frac{1}{2}\sin(2t) \right) + C\\
= & -\frac{1}{2}\left(t - \sin(t)\cos(t) \right) + C, ~ \sin(2t) = 2\sin(t)\cos(t)\\
= & -\frac{1}{2}\left(\cos^{-1}(x) - x\sqrt{1-x^{2}}\right) + C
\end{align*}

So that got tough...  As the book notes, we are in effect doing an *inverse substitution*, which is why we have to think of the substitution as simultaneously

$$
x = \cos(t), ~ t = \cos^{-1}(x)
$$

but then we also need to remember

$$
\sin(t) = \sqrt{1-\cos^{2}(t)} = \sqrt{1-x^{2}}.
$$

So yeah, it's kind of all the things at once.  

And now let's make it a bit weirder and ask how to find 

$$
\int\sqrt{a^{2}-x^{2}}dx = a \int\sqrt{1 - \frac{x^{2}}{a^{2}}}dx
$$

So I now use the inverse substitution $x = a\cos(t)$.  This then gets me the even more complicated answer 

\begin{align*}
\int\sqrt{a^{2}-x^{2}}dx = & -a^{2}\int\sin^{2}(t)dt \\
=& -\frac{a^{2}}{2}\left(\cos^{-1}\left(\frac{x}{a}\right) - \frac{x}{a}\sqrt{1-\frac{x^{2}}{a^{2}}} \right) + C
\end{align*}

Note, here we have to use 

$$
x = a\cos(t), ~ t = \cos^{-1}\left(\frac{x}{a} \right).
$$

**Problem**: Find 

$$
\int \frac{\sqrt{9-x^{2}}}{x^{2}}dx
$$

So our friend here is $x = \sqrt{9}\cos(t) = 3\cos(t)$.
