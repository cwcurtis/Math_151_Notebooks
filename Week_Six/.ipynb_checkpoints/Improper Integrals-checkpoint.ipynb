# 7.8 Improper Integrals

## Infinite Intervals

Suppose we looked at the following function, $f(t;p)$ defined by an integral:

$$
f(t;p) = \int_{1}^{t}\frac{dx}{x^{p}}, ~ p \gt 0.
$$

Note, we can readily see that 

$$
\lim_{x\rightarrow \infty} \frac{1}{x^{p}} = 0, ~ p \gt 0,
$$

and so we see that the integrand vanishes in the infinite limit for all positive values of the power $p$.  Now we can ask, how much area is underneath these curves defined on an infinite interval?  

To answer that, of course we can readily show that for $p\neq 1$ we have 

\begin{align*}
f(t;p) = & \left.\frac{x^{-p+1}}{-p+1}\right|_{1}^{t} \\
= & \frac{t^{1-p}-1}{1-p} \\
= & \frac{1-t^{1-p}}{p-1}
\end{align*}

and if $p=1$, then we have that 

$$
f(t;1) = \left.\ln|x|\right|_{1}^{t} = \ln|t| - \ln 1 = \ln t
$$


Now we can ask about the following limit

$$
\lim_{t\rightarrow \infty}f(t;p).
$$

* Show that $\lim_{t\rightarrow \infty} f(t;p) = \infty$ if $0 \lt p \leq 1$.
* Show that $\lim_{t\rightarrow \infty} f(t;p) = \frac{1}{p-1}$ if $p \gt 1$.

So for the $p\gt 1$ case, we now say that the improper integrals are **convergent** and 

$$
\int_{1}^{\infty}\frac{dx}{x^{p}} = \lim_{t\rightarrow \infty} \int_{1}^{t}\frac{dx}{x^{p}} = \frac{1}{p-1}.  
$$

If $0\lt p \leq 1$, we then say that the improper integrals are **divergent** and 
$$
\int_{1}^{\infty}\frac{dx}{x^{p}} = \infty.
$$

Incidentally, we now see that for $p\gt 1$, there is a finite amount of area of the under curve, but for $0\lt p \leq 1$, there is an inifinte amount of area, even though the integrand in every case vanishes to zero in the infinite limit.  

**Problem**: Find 
$$
\int_{-\infty}^{0}xe^{x}dx.
$$
Technically, we need to rewrite this as 
$$
\int_{-\infty}^{0}xe^{x}dx = \lim_{t\rightarrow \infty}\int_{-t}^{0}xe^{x}dx.
$$
After that, integration-by parts is your friend.
