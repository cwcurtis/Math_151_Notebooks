# 7.3 - Trig Substitutions 

What to do with integrals of the form 

$$
\int \sqrt{1-x^{2}} dx, ~ \int \sqrt{1+x^{2}} dx, ~ \int \sqrt{x^{2}-1} dx
$$

Well, we have three fundamental identities

$$
1-\cos^{2}(t) = \sin^{2}(t), ~ 1+ \tan^{2}(t)= \sec^{2}(t), ~ \sec^{2}(t)-1=\tan^{2}(t)
$$

So in the first case, let us use $x = \cos(t)$ so that $dx = -\sin(t)dt$ and we have 

\begin{align*}
\int \sqrt{1-x^{2}} dx = & -\int \sqrt{1-\cos^{2}(t)}\sin(t) dt\\
= & -\int \sin^{2}(t) dt \\
= & -\frac{1}{2}\int \left( 1 - \cos(2t)\right) dt\\
= & -\frac{1}{2}\left(t - \frac{1}{2}\sin(2t) \right) + C\\
= & -\frac{1}{2}\left(t - \sin(t)\cos(t) \right) + C, ~ \sin(2t) = 2\sin(t)\cos(t)\\
= & -\frac{1}{2}\left(\cos^{-1}(x) - x\sqrt{1-x^{2}}\right) + C
\end{align*}

So that got tough...  As the book notes, we are in effect doing an *inverse substitution*, which is why we have to think of the substitution as simultaneously

$$
x = \cos(t), ~ t = \cos^{-1}(x)
$$

but then we also need to remember

$$
\sin(t) = \sqrt{1-\cos^{2}(t)} = \sqrt{1-x^{2}}.
$$

So yeah, it's kind of all the things at once.  

And now let's make it a bit weirder and ask how to find 

$$
\int\sqrt{a^{2}-x^{2}}dx = a \int\sqrt{1 - \frac{x^{2}}{a^{2}}}dx
$$

So I now use the inverse substitution $x = a\cos(t)$.  This then gets me the even more complicated answer 

\begin{align*}
\int\sqrt{a^{2}-x^{2}}dx = & -a^{2}\int\sin^{2}(t)dt \\
=& -\frac{a^{2}}{2}\left(\cos^{-1}\left(\frac{x}{a}\right) - \frac{x}{a}\sqrt{1-\frac{x^{2}}{a^{2}}} \right) + C
\end{align*}

Note, here we have to use 

$$
x = a\cos(t), ~ t = \cos^{-1}\left(\frac{x}{a} \right).
$$

**Problem**: Find 

$$
\int \frac{\sqrt{9-x^{2}}}{x^{2}}dx
$$

So our friend here is $x = \sqrt{9}\cos(t) = 3\cos(t)$.

**Problem**: Moving on, we can ask, find the area of an ellipse defined by the equation 


$$
\frac{x^{2}}{a^{2}} + \frac{y^{2}}{b^{2}} = 1.
$$

So to do this, we note that if we solve for $y$ so that 

$$
y = b\sqrt{1-\frac{x^{2}}{a^{2}}}
$$

Then we see that the area $A$ is given by 

$$
A = 2b\int_{-a}^{a}\sqrt{1-\frac{x^{2}}{a^{2}}} dx = 4b \int_{0}^{a}\sqrt{1-\frac{x^{2}}{a^{2}}} dx
$$

Okay, using $x=a\cos(t)$ take it away...  remember, $\cos(\pi/2)=0$ and $\cos(0)=1$, and yeah, you're gonna want to change the bounds.  

**Problem**: Find 

$$
\int \frac{dx}{x^{2}\sqrt{x^{2}+4}}
$$

So here, we need a different approach.  Working from 

$$
1 + \tan^{2}(t) = \sec^{2}(t)
$$

then we're doing to use $x = \sqrt{4}\tan(t) = 2\tan(t)$.  Then we have that $dx = 2\sec^{2}(t) dt$ and thus 

\begin{align*}
\int \frac{dx}{x^{2}\sqrt{x^{2}+4}} = & \int \frac{2\sec^{2}(t) }{4\tan^{2}(t)\sqrt{4\tan^{2}(t)+4}} dt\\
= & \frac{1}{4}\int \frac{\sec^{2}(t) }{\tan^{2}(t)\sec(t)} dt\\
= & \frac{1}{4}\int \frac{\sec(t) }{\tan^{2}(t)} dt\\
= & \frac{1}{4}\int \frac{\cos(t) }{\sin^{2}(t)} dt
\end{align*}

See if you can get it the rest of the way...

**Problem**: Find

$$
\int \frac{dx}{\sqrt{x^{2}-a^{2}}}
$$

So here, we note that 

$$
\sec^{2}(t) - 1 = \tan^{2}(t)
$$

so we use the substitution $x = a\sec(t)$, so that $dx = a \sec(t)\tan(t)dt$ and 

\begin{align*}
\int \frac{dx}{\sqrt{x^{2}-a^{2}}} = & \int \frac{a\sec(t)\tan(t)}{\sqrt{a^{2}\sec^{2}(t)-a^{2}}}dt\\
= & \int \frac{\sec(t)\tan(t)}{\tan(t)}dt\\
= & \int \sec(t) dt\\
= & \int \sec(t)\left(\frac{\sec(t)+\tan(t)}{\sec(t)+\tan(t)}\right) dt, ~ \text{TRICK!!!!!!}
\end{align*}

Take it from there...

**Problem**: Find

$$
\int_{0}^{3\sqrt{3}/2} \frac{x^{3}}{\sqrt{4x^{2}+9}} dx = \frac{1}{2}\int_{0}^{3\sqrt{3}/2} \frac{x^{3}}{\sqrt{x^{2}+\frac{9}{4}}} dx
$$

So, we see that $x = \frac{3}{2}\tan(t)$ is the weapon of choice here.  See if you can take it from here.

**Problem**:  Okay but what about

$$
\int \frac{x}{\sqrt{3-2x-x^{2}}} dx
$$
    
    
So the trick here is to complete the square so that 

\begin{align*}
3-2x-x^{2} = & 3 -\left(x^{2}+2x\right) \\
= & 3 - \left(x^{2}+2x + 1 - 1\right) \\
= & 3 - \left((x+1)^{2} - 1\right) \\ 
= & 4 - \left(x+1\right)^{2}
\end{align*}

Then our integral becomes 

$$
\int \frac{x}{\sqrt{4 - \left(x+1\right)^{2}}} dx
$$

and so we use the substitution $x+1 = 2\cos(t)$.  Take it from there...

# 7.4 - Partial Fractions

So, we are now going to look at integrating rational functions, which is to say we are going to study integrals of the form 

$$
\int \frac{P(x)}{Q(x)}dx, ~ P(x), ~Q(x) ~\text{are polynomials}.  
$$

To do this, we are going to need, you guessed it, more trickery.  To wit, let's look at 

$$
\int \frac{dx}{x^{2}-1},
$$

Note, this corresponds to the choices 

$$
P(x)=1, ~ Q(x) = x^{2}-1
$$

Now, we note that 

$$
\int \frac{dx}{x^{2}-1} = \int \frac{dx}{(x+1)(x-1)}
$$

Here comes the trick.  We ask, hey, what if we tried the following decomposition

$$
\frac{1}{(x+1)(x-1)} = \frac{A}{x+1} + \frac{B}{x-1}.   
$$

This is called a **partial fractions** decomposition.  Now why do we want this?  Well, if we had it, it makes find the integral easy since 

\begin{align*}
\int \frac{dx}{x^{2}-1} = & \int \frac{dx}{(x+1)(x-1)} \\
= & \int \left(\frac{A}{x+1} + \frac{B}{x-1} \right) dx\\
= & A\ln|x+1| + B\ln|x-1| + C
\end{align*}


So now I just need to find $A$ and $B$.  How?  Well, we see via cross multiplication that 

$$
\frac{A}{x+1} + \frac{B}{x-1} = \frac{A(x-1) + B(x+1)}{(x+1)(x-1)}
$$

so if we want this to equal the original expression, we match numerators so that 

$$
A(x-1) + B(x+1) = 1.
$$

> NOTE: THIS EXPRESSION MUST HOLD FOR ALL POSSIBLE x!!!!!!!!!!!!!!!!!!!!!!!!!!!

So if I choose $x=1$, the equation becomes $2B = 1$, so $B=1/2$.  Likewise, if I choose $x=-1$, then we get $A=-1/2$ so that 

$$
\frac{1}{x^{2}-1} = -\frac{1}{2}\frac{1}{(x+1)} + \frac{1}{2}\frac{1}{(x-1)}
$$

and the integral is given by 

$$
\int \frac{dx}{x^{2}-1} = -\frac{1}{2}\ln|x+1| + \frac{1}{2}\ln|x-1| + C = \frac{1}{2}\ln\left|\frac{x-1}{x+1} \right| + C
$$


Now, let's bump this up a bit 

**Problem**: Find 
$$
\int \frac{x^{3}+2}{x^{2}-1} dx
$$

So, the first thing we want to do is break this up into refined pieces.  To do this, we are going to use synthetic division.  This is really hard to type, so I'm just going to write the result in the notes. 

$$
x^{3} + 2 = x(x^{2}-1) + x + 2
$$
 
We then see that 

$$
\int \frac{x^{3}+2}{x^{2}-1} dx = \int \left(x + \frac{x}{x^{2}-1} + \frac{2}{x^{2}-1} \right) dx
$$

Take it away now... 

So that was slick.  Now, we need a little bit of theory to figure out how to push this approach even further.  First, we define the degree of a polynomial.  

> **Definition**: The degree of a polynomial $Q(x)$ is the value of its highest power.  We denote this as $\text{deg}(Q)$.

So okay, if 

$$
Q(x) = 4x^{3} + 2x^{2} -2x + 1, 
$$

then

$$
\text{deg}(Q) = 3
$$

since $x^{3}$ is the highest power.  The other thing we want to mess with here is the idea of a **monomial**.  A polynomial $Q(x)$ is a monomial if the coefficient multiplying the polynomial term of highest power is just one.  So using the example above, we have 

$$
Q(x) = 4x^{3} + 2x^{2} -2x + 1 = 4\left(x^{3} + \frac{1}{2}x^{2} - \frac{1}{2}x + \frac{1}{4} \right)
$$

so while $Q$ is **not** a monomial (because of the $4x^{3}$ term), by factoring out that $4$, we get that $Q$ can be written as the product of $4$ and the monomial 

$$
x^{3} + \frac{1}{2}x^{2} - \frac{1}{2}x + \frac{1}{4}.
$$

So moral of story, any polynomial is already a monomial or I can turn it into one just by factoring out a coefficient.  Okay, now why is that important?  Because we can factor monomials by roots.  So for example

$$
2x^{2} - 2 = 2(x^{2}-1) = 2(x-1)(x+1).
$$

Another example

$$
3x^{2} - 6x + 3 = 3(x^{2} - 2x + 1) = 3(x-1)^{2}.
$$

Note, in the first example we have two **distinct** roots, $x=\pm 1$.  In the second, we have a **repeated** root $x=1$.  Moreover, note that in both cases the degree of the polynomial is $2$.  While we cannot get into this, there are no accidents in math, and you should generally know

> **Fundamental Theorem of Algebra**: For a polynomial $Q(x)$ such that $\text{deg}(Q)=n$, it is has $n$ roots, counting repetitions.  

So in both examples, we have polymoials of degree 2, and both have two roots.  In the first, the roots are distinct, in the second case, they are repeated.  Now as a side note, we note that the Fundamental Theorem of Algebra does not say $n$ **real** roots. To wit

$$
x^{2} + 1 = (x+i)(x-i), ~ i=\sqrt{-1}.
$$

Polynomials with complex roots are called **irreducible**.  Note, for the general quadratic equation
$$
ax^{2} + bx + c = 0,
$$
we know the roots are given by the formula 
$$
x_{\pm} = \frac{1}{2a}\left(-b\pm \sqrt{b^{2}-4ac} \right)
$$
Thus, we see a quadratic expression will be irreducible if $b^{2}-4ac \lt 0$, and we will get repeated roots if $b^{2}-4ac = 0$.  

So now let's generalize our approach to integrals of the form 

$$
\int\frac{P(x)}{Q(x)}dx, ~ \text{deg}(Q(x)) = n, ~ \text{deg}(P(x)) \lt \text{deg}(Q(x))
$$

### Case I: $Q(x)$ is reducible.  

* Step 1: Write $Q(x)=c\tilde{Q}(x)$ where $\tilde{Q}(x)$ is a monomial.  
* Step 2: Factor $\tilde{Q}(x)$ by roots so that 
$$
\tilde{Q}(x) = (x-x_{1})^{p_{1}}\cdots (x-x_{j})^{p_{j}}, ~ p_{1} + \cdots + p_{j} = n.
$$

#### Case 1a: No Repeated Roots:
Now if none of the factors are repeated, i.e. $p_k=1$ in all cases or we have $n$ distinct roots, then we just let 

$$
\frac{P(x)}{c\tilde{Q}(x)} = \frac{C_{1}}{x-x_{1}} + \frac{C_{2}}{x-x_{2}} + \cdots + \frac{C_{n}}{x-x_{n}}.
$$

Through cross-multiplication we must have that

$$
C_{1}(x-x_{2})\cdots(x-x_{n}) + C_{2}(x-x_{1})(x-x_{3})\cdots(x-x_{n}) + \cdots + C_{n}(x-x_{1})\cdots(x-x_{n-1}) = \frac{P(x)}{c}
$$

and thus we see to find $C_{1}$ that we would choose $x=x_{1}$ so that every term zeros out except for the first and thus we find 

$$
C_{1} = \frac{P(x_{1})}{c(x_{1}-x_{2})\cdots(x_{1}-x_{n})}.
$$

Likewise we have 

$$
C_{2} = \frac{P(x_{2})}{c(x_{2}-x_{1})(x_{2}-x_{3})\cdots(x_{2}-x_{n})},
$$

and this proceeds till we finally find that 

$$
C_{n} = \frac{P(x_{n})}{c(x_{n}-x_{1})(x_{n}-x_{2})\cdots(x_{n}-x_{n-1})}.
$$


**Example**: 
$$
\frac{P(x)}{Q(x)} = \frac{x^{2}-2}{x^{3}-x} = \frac{x^{2}-2}{x(x-1)(x+1)}
$$


So, we then let 

$$
\frac{x^{2}-2}{x^{3}-x} = \frac{C_{1}}{x} + \frac{C_{2}}{x-1} + \frac{C_{3}}{x+1}
$$

Then we have 

$$
C_{1}(x-1)(x+1) + C_{2}x(x+1) + C_{3}x(x-1) = x^{2} - 2
$$

So, starting with $x=0$, we get 

$$
-C_{1} = -2, ~ C_{1} = 2.
$$

then $x=1$, and we get 

$$
2C_{2} = -1, ~ C_{2} = -1/2
$$

and then for $x=-1$ we get 

$$
2C_{3} = -1, ~ C_{3} = -1/2
$$

and thus we find that 

$$
\frac{x^{2}-2}{x^{3}-x} = \frac{2}{x} -\frac{1}{2} \frac{1}{(x-1)} - \frac{1}{2} \frac{1}{(x+1)}
$$

#### Case 1b: Repeated Roots:

So this is better explained via an example.  

**Example**: 
$$
\frac{P(x)}{Q(x)} = \frac{1}{x(x-1)^{2}}
$$

So in this case, we need to let 

$$
\frac{1}{x(x-1)^{2}} = \frac{C_{1}}{x} + \frac{C_{21}}{x-1} + \frac{C_{22}}{(x-1)^{2}}
$$

Cross multiplication then gets us 

$$
C_{1}(x-1)^{2} + C_{21}x(x-1) + C_{22}x = 1
$$

Now since we have two roots, $x=0$ and $x=1$, we immediately get that 

$$
C_{1} = 1, ~ C_{22}=1.
$$

Now, probably the easiest thing to do at this point is just pick another point $x$ different from $0$ and $1$.  So, let's choose $x=2$, so that our equation becomes

$$
1 + 2C_{21} + 2 = 1, ~ C_{21} = -1,
$$

and thus 

$$
\frac{1}{x(x-1)^{2}} = \frac{1}{x} - \frac{1}{x-1} + \frac{1}{(x-1)^{2}}
$$


To generalize, we see that if in our factorization of $\tilde{Q}$ we have a term $(x-x_{l})^{p_{l}}$, then in the partial fractions decomposition, we need to include a sequence of terms of the form 

$$
\frac{C_{l1}}{x-x_{l}} + \frac{C_{l2}}{(x-x_{l})^{2}} + \cdots + \frac{C_{lp_{l}}}{(x-x_{l})^{p_{l}}}
$$

So, for example, we would do the following for $P(x)/Q(x) = (x^{2}-4)/(x(x-1)^{3}(x-2)^{4})$ 

$$
\frac{x^{2}-4}{x(x-1)^{3}(x-2)^{4}} = \frac{C_{1}}{x} + \frac{C_{21}}{x-1} + \frac{C_{22}}{(x-1)^{2}} + \frac{C_{23}}{(x-1)^{3}} + \frac{C_{31}}{x-2} + \frac{C_{32}}{(x-2)^{2}} + \frac{C_{33}}{(x-2)^{3}} + \frac{C_{34}}{(x-2)^{4}}
$$

### Case II: $Q(x)$ has an irreducible factor.

So let's keep in mind that the canonical case of an irreducible factor comes up when we try to find the integral

$$
\int \frac{dx}{x^{2} + a^{2}} = \frac{1}{a}\tan^{-1}\left(\frac{x}{a} \right) + C
$$

So in many ways, the theory we developed above carries over, we just need to add a few tweaks.  For example, say we wanted to find 

$$
\int\frac{2x^{2}-1}{x(x^{2}+1)}dx.
$$

To account for the irreducible factor, we use the following partial fractions decomposition

$$
\frac{2x^{2}-1}{x(x^{2}+1)} = \frac{C_{1}}{x} + \frac{C_{21}x + C_{22}}{x^{2} + 1}
$$

To figure this out, we see that after cross multiplication that we have 

$$
C_{1}(x^{2}+1) + x(C_{21}x + C_{22}) = 2x^{2} - 1 
$$

So, in this case, it actually makes more sense to distribute and collect powers.  By that I mean that on the left hand side, we rewrite the expression as 

$$
(C_{1}+C_{21})x^{2} + C_{22}x + C_{1} = 2x^{2} - 1 
$$

and then we match so that 

$$
C_{1} + C_{21} = 2, ~ C_{22} = 0, ~ C_{1} = -1.
$$

We quickly see then that $C_{21}=3$ and finally we have 

$$
\frac{2x^{2}-1}{x(x^{2}+1)} = -\frac{1}{x} + \frac{3x}{x^{2} + 1}
$$


Thus

$$
\int\frac{2x^{2}-1}{x(x^{2}+1)}dx = \int \left(-\frac{1}{x} + \frac{3x}{x^{2} + 1} \right) dx = -\ln|x| + \frac{3}{2}\ln|x^{2}+1| + C
$$

Likewise, we could of course look at repeated irreducible factors, like in the following problem

$$
\int\frac{2x^{2}-1}{x(x^{2}+1)^{2}}dx
$$

Here, we would use the partial-fraction decomposition

$$
\frac{2x^{2}-1}{x(x^{2}+1)^{2}} = \frac{C_{1}}{x} + \frac{C_{21}x + C_{22}}{x^{2}+1} + \frac{C_{23}x + C_{24}}{(x^{2}+1)^{2}}.
$$

Cross multiplication and carefully expanding and collecting by powers gets you the following equation

$$
(C_{1}+C_{21})x^{4} + C_{22}x^{3} + (2C_{1}+C_{21}+C_{23})x^{2} + (C_{22}+C_{24})x + C_{1} = 2x^{2} - 1
$$

Matching by powers on the left and right, we get that

$$
C_{1} + C_{21} = 0, ~ C_{22} = 0, ~ 2C_{1} + C_{21} + C_{23} = 2, ~ C_{22} + C_{24} = 0, ~ C_{1} = -1.
$$

After a bit of balancing, we see that $C_{21} = 1$, $C_{24}=0$, ~ $C_{23}=3$ and thus 

$$
\frac{2x^{2}-1}{x(x^{2}+1)^{2}} = -\frac{1}{x} + \frac{x}{x^{2}+1} + \frac{3x}{(x^{2}+1)^{2}}.
$$
