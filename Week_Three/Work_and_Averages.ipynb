## 6.4 - Work

So, let's dabble in some physics.  Now, Newton is famous for saying few different things about *force*.  And what is force?  Well it's the thing that makes something that was at rest no longer at rest... among other things.  Mathematically, force, say $F$ is encapsulated in Newton's 2nd Law which says that 

$$
F = ma = m \frac{d^{2}x}{dt^{2}}.
$$

The most well known force most of us experience is that of gravity, which we describe at the surface of the Earth via the equation 

$$
F = mg, ~ g = 9.8 m/s^{2}
$$

If we measure $m$ is kilograms, then $F$ is measured in Newtons (not the cookies).  Likewise, we also talk about the force of friction, which gets us to the next point.  

When you try to push things over surfaces which provide resistance via friction, we often notice that we get tired of doing it.  This is because we expend energy or do work.  Work for our purposes will be defined to be 

$$
W = Fd
$$

where $d$ is how far we've displaced something while applying a force $F$.  Now of course, what if the force is varying as we move it along a given distance?  Well, sure enough, we then measure $dW = F(x)dx$ so that the net work along a given path is given by 

$$
W = \int_{a}^{b}F(x)dx.  
$$

### Hooke's Law  

So, as some of you may have encountered in physics, if we are interested in how much work a spring does as it moves back and forth, we need to use *Hooke's Law* which describes the force the spring applies back as it is pushed or pulled.  

![](hooke.png)

So here 

$$
F(x) = kx
$$


So the work done as the spring is pulled into extension from equilibrium position $x=0$ to say the point $x=a$ is 

$$
W_{0,a} = \int_{0}^{a}kx dx = \left.\frac{k}{2}x^{2}\right|_{0}^{a} = \frac{ka^{2}}{2}.
$$

**Problem**: How much work is required to extend this spring 1.8 m?  

### Lifting Chains

So, suppose I think about a chain of length $L$ hanging down the side of a wall (or a building as the book does).  It starts with the whole chain hanging down.  How much work do I perform by lifting the chain against gravity?  

So this problem is hard, because if you think about it, as I pull parts of the chain up onto the wall, the amount of mass I move decreases (assuming we let the parts we've pulled up to rest on the ground).  Right, there is just less chain to lift as I keep lifting it.  So to address this mathematically, I need to think, okay, now the mass is evenly distributed through the chain (not too unreasonable of an assumption).  So if 

* Whole chain is mass $M$
* Length of chain is $L$, so that means we have $M/L$ mass distributed over the whole length.  
* If we have already pulled up $x$ length of chain, then we have $L-x$ length remaining, and we have $M(L-x)/L$ mass of chain remaining.  

So now, the differential of work will be 

$$
dW = g\frac{M}{L}(L-x)dx
$$

and the total work done in lifting the chain is 

\begin{align*}
W = & \frac{Mg}{L}\int_{0}^{L}\left(L-x \right)dx \\
= & \frac{Mg}{L}\left.\left(Lx - \frac{x^2}{2} \right)\right|_{0}^{L} \\
= & \frac{MgL}{2}.
\end{align*}

So if you'll note, this is where the book and I diverge.  It keeps showing you how to solve one problem at a time.  I'm showing you how to solve any problem of a given form and then plug numbers in at the end.  Said another way

> Calculus gives us physically important formulas. Compute once, move on with your life!!

## 6.5 - Averages 

So, say I have a function $f(x)$ on the interval $[a,b]$.  What is its *average* value?  Well, if we subdivided $[a,b]$ into $N$ intervals of equal width $\Delta x = \frac{b-a}{N}$, then we could approximate the average of $f$, say $\left<f\right>$, via the Riemann sum

$$
\left<f\right> \approx \frac{1}{N}\left(f(x_{1})+\cdots +f(x_{N}) \right) = \frac{1}{b-a}\sum_{j=1}^{N}f(x_{j})\Delta x
$$

So, doing the usual thing, we then get as $N\rightarrow \infty$ ($\Delta x\rightarrow 0$) that 

$$
\left<f\right> = \frac{1}{b-a}\int_{a}^{b}f(x) dx.
$$

**Problem**: Show that $\left<\cos(x)\right> = \left<\sin(x)\right> = 0$ for $0\leq x \leq 2\pi$

### The Mean-Value Theorem for Integrals 

There is a funny little theorem affiliated with averages, and it says:

> For $f(x)$ continous on $a\leq x \leq b$, there exists some value $c$ such that $a\lt c \lt b$ and 
$$
f(c) = \frac{1}{b-a}\int_{a}^{b}f(x) dx.  
$$

So every continuous function takes on its average over a given interval.  Who knew?  

# 7.1 - Integration-by Parts 

So, let me teach you one of the most useful results in Calculus ever.  If we think back, we should remember the product rule for differentiation

$$
\frac{d}{dx}(uv) = v(x)\frac{d}{dx}u(x) + u(x)\frac{d}{dx}v(x)
$$

Now what if I integrated all of this with respect to $x$?

$$
\int \frac{d}{dx}(uv) dx = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

Well, using the Fundamental Theorem, the left-hand side  of all of this must give us 

$$
u(x)v(x) + C = \int v(x)\frac{d}{dx}u(x) dx + \int u(x)\frac{d}{dx}v(x) dx
$$

The way we are going to write this though is 

$$
\int u(x)\frac{d}{dx}v(x) dx = u(x)v(x) - \int v(x)\frac{d}{dx}u(x) dx
$$

leaving the constant of integration $C$ to be implied.  This formula sets up a technique called *integration-by parts*.

Let's see how it's used.  

**Problem**: Find $\int x\sin(x) dx$.  So the hard thing to do here is identify a $u(x)$ and a $\frac{dv}{dx}$.  Were I you, I would let 

$$
u(x) = x, ~ \frac{dv}{dx} = \sin(x).
$$

Now fine... this means $v(x) = -\cos(x)$ (ignore the constant of integration for now...)  Following the formula from above we then have 

\begin{align*}
\int x\sin(x) dx = & -x\cos(x) - \int \left(-\cos(x)\right) \left(\frac{d}{dx} x\right) dx  \\ 
 = & -x\cos(x) + \int \cos(x) dx  \\ 
 = & -x\cos(x) + \sin(x) + C
\end{align*}

What just happened?  I know, it's weird.  So the idea is the following 

* We have a hard integral
* But the hard integral is the product of two functions, either of which is easy to integrate.
* And if we integrate the "correct" function, we end up differentiating the other, and differentiation has a way of making things get simpler $(d x^{n} /dx = nx^{n-1})$

So okay, you're going, okay, but what?  How do I pick?  So let's do the same problem, but let's see if we had tried the other way.  So now we choose  

$$
u(x) = \sin(x), ~ \frac{dv}{dx} = x.
$$

Okay, that means $v(x) = x^{2}/2$ and we then get 

$$
\int x\sin(x) dx = \sin(x)\frac{x^{2}}{2} - \frac{1}{2}\int x^{2}\cos(x) dx 
$$

So we see, that did not get easier, it got harder.  Now we see why the other choice was better.  

**Problem**: Find 
$$
\int \ln(x) dx
$$

So the trick here (and this is such a tricky little trick) is to write 

$$
\int \ln(x) dx = \int \left(\ln(x) \times 1 \right)dx
$$

we then make the choice 

$$
u(x) = \ln(x), ~ \frac{dv}{dx} = 1 \rightarrow v(x) = 1
$$

Using our formula then, we find that 

\begin{align*}
\int \ln(x) dx = & x\ln(x) - \int x \frac{d}{dx}\ln(x) dx\\
= & x\ln(x) - \int x \frac{1}{x} dx\\
= & x\ln(x) - \int dx\\
= & x\ln(x) - x + C
\end{align*}
